```python
## NUMBER 1.HANDLING UNKNOWNS
```

```python
'''
Checking columns with unknown values and how many there are in the dataset
'''
bank=bank_data.replace('unknown',np.nan)
missing_value_check=bank.isna().sum()
missing_value_check
```

```python
print("""The strategy i will apply to handle the missing values is by logically checking the
correlations and relationships between the columns and filling it with the possible 
correct value.""")
```

```python
'''
filling the job with the possible correct values:
when the clients age above 60yrs, they are likely to be retired
'''
bank.loc[(bank["age"]>60)&(bank["job"].isna()), "job"]='retired'
bank.isna().sum()
```

```python
'''
filling the missing value in default with the most likely value:
when the client has no personal loan, the default credit is likely to be no
'''
bank.loc[(bank["loan"]=="no")&(bank["default"].isna()), "default"]='no'
bank.isna().sum()
```

```python
'''
when the client has a housing loan and a personal loan, the default credit is likely to be a yes
'''
bank.loc[(bank["housing"]=="yes")&
    (bank["loan"]=="yes")&
    (bank["default"].isna()), "default"]="yes"
bank.isna().sum()
```

```python
'''
filling the loan; when default is no,personal loan is likely to be no
'''
bank.loc[(bank["default"]=="no")&
    (bank["loan"].isna()), "loan"]="no"
bank.isna().sum()
```

```python
"""
filled missing values possible, now drop the rest missing values because missing data is below 30% of the dataset, NOW DATA HAS NO MISSING VALUES
"""
bank_cleansed=bank.dropna()
bank_cleansed
```

```python
## NUMBER 1. ENCODING
```

```python
print("Features to be encoded include: 'job', 'marital', 'education', 'default',\n'housing', 'loan', 'day_of_week', 'month','poutcome','y'")
```

```python
'''
Encoding atleast three features using pandas; job, marital, education, default, housing, loan, month, poutcome
'''
encoded=pd.get_dummies(bank_cleansed,columns=['job', 'contact', 'marital', 'education', 'default', 'housing', 'loan', 'month','poutcome',], drop_first=True)
encoded
```

```python
## NUMBER 2. TARGET ENCODING
```

```python
'''
Encoding the target variable y: 1 for yes; 0 for no
'''
encoded['y']=encoded['y'].replace({'yes':1, 'no':0})
```

```python
## NUMBER 3. FEATURE SCALING
```

```python
'''
The Feature in the dataset that requires scaling
'''
print("THE FEATURE THAT REQUIRES SCALING IN THE DATASET IS 'nr.employed', 'pdays'. Because it has way\n higher values compared to the other features so that the model doesnt treat it more important")
```

```python
'''
Standardization to the relevant numerical features
'''
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
encoded['standardized_nr.employed']=scaler.fit_transform(encoded[['nr.employed']])
```

```python
encoded['standardized_pdays']=scaler.fit_transform(encoded[['pdays']])
```

```python
## NUMBER 4. FEATURE ENGINEERING
```

```python
'''
ARE THERE FEATURE I WOULD CONSIDER TO DROP???
'''
print("YES!!! I would like to drop some features that are highly correlated regarding insight from the heatmap,\n features such as 'emp.var.rate' and 'euribor3m' with each other and with 'nr.employed' \n Irrelevant feature like 'day_of_the_week' and 'contact' should be\n dropped as well because they aint important for prediction")
```

```python
p=encoded.drop(['emp.var.rate','euribor3m','day_of_week'], axis=1)
bankencoded=p.drop(columns=[col for col in p.columns if col.startswith(('month_') or ('contact_'))])
bankencoded
```

```python
'''
creating new features to help prediction
'''
bankencoded['persuasion']=bankencoded['duration']+bankencoded['campaign']
```
